```python
enjoySport = pd.read_csv('/kaggle/input/enjoysport/ENJOYSPORT.csv', encoding='latin')
enjoySport
```

```python
concepts = np.array(enjoySport.iloc[:,0:-1])
print("The attributes are: ",concepts)
```

```python
target = np.array(enjoySport.iloc[:,-1])
print("\n The target is: ",target)
```

```python
def learn(concepts, target):
    specific_h = concepts[0].copy()
    print("\n Initialization of specific_h and general_h")
    print(specific_h)
    general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print(general_h)
    for i,h in enumerate(concepts):
        if target[i] == 1:
            for x in range(len(specific_h)):
                if h[x]!= specific_h[x]:
                    specific_h[x] ='?'
                    general_h[x][x] ='?'
                print(specific_h)
            print(specific_h)
        if target[i] == 0:
            for x in range(len(specific_h)):
                if h[x]!= specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = '?'
        print("\n Steps of Candidate Elimination Algorithm",i+1)
        print(specific_h)
        print(general_h)
    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?', '?']]
    for i in indices:
        general_h.remove(['?', '?', '?', '?', '?', '?'])
    return specific_h, general_h
```

```python
s_final, g_final = learn(concepts, target)
```

```python
print("\nFinal Specific_h:", s_final, sep="\n")
print("\nFinal General_h:", g_final, sep="\n")
```

```python
second_data = pd.read_csv('/kaggle/input/purchase/purchase.csv', encoding='latin')
second_data
```
